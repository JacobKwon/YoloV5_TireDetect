```python
import pandas as pd
```

```python
base_path = './car_data/car_label'

train = pd.read_csv(base_path+'/train/_annotations.csv')
test = pd.read_csv(base_path+'/test/_annotations.csv')
val = pd.read_csv(base_path+'/valid/_annotations.csv')
train.head()
```

```python
from sklearn import preprocessing, model_selection

le = preprocessing.LabelEncoder()
le.fit(train['class'])
print(le.classes_)

# labels = le.transform(train_df['class'])
# df['labels'] = labels
# print(labels)
```

```python
import cv2 as cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

img = cv2.imread(base_path + '/train/' + train['filename'][0])
print(img.shape)
# plt.axis([train_df['xmin'][0], train_df['xmax'][0], train_df['ymin'][0], train_df['ymax'][0]])  # X, Y축의 범위: [xmin, xmax, ymin, ymax]
# plt.imshow(img)

points = np.array([
                    [train['xmax'][0], train['ymin'][0]],
                    [train['xmin'][0], train['ymin'][0]],
                    [train['xmin'][0], train['ymax'][0]],
                    [train['xmax'][0], train['ymax'][0]]
                  ])

points2 = np.array([
                    [train['xmax'][1], train['ymin'][1]],
                    [train['xmin'][1], train['ymin'][1]],
                    [train['xmin'][1], train['ymax'][1]],
                    [train['xmax'][1], train['ymax'][1]]
                  ])

# image = cv2.polylines(img, [points], True, (0, 0, 255), 4)
image = cv2.polylines(img, [points,points2], True, (0, 0, 255), 4)

plt.imshow(image)
plt.show()
```

```python
fig = plt.figure()

#add axes to the image
ax = fig.add_axes([0,0,1,1])

# read and plot the image
image = plt.imread('images/1.jpg')
plt.imshow(image)

# iterating over the image for different objects
for _,row in train[train.image_names == "1.jpg"].iterrows():
    xmin = row.xmin
    xmax = row.xmax
    ymin = row.ymin
    ymax = row.ymax
    
    width = xmax - xmin
    height = ymax - ymin
    
    # assign different color to different classes of objects
    if row.cell_type == 'RBC':
        edgecolor = 'r'
        ax.annotate('RBC', xy=(xmax-40,ymin+20))
    elif row.cell_type == 'WBC':
        edgecolor = 'b'
        ax.annotate('WBC', xy=(xmax-40,ymin+20))
    elif row.cell_type == 'Platelets':
        edgecolor = 'g'
        ax.annotate('Platelets', xy=(xmax-40,ymin+20))
        
    # add bounding boxes to the image
    rect = patches.Rectangle((xmin,ymin), width, height, edgecolor = edgecolor, facecolor = 'none')
    
    ax.add_patch(rect)
```
